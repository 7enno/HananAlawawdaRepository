## 4. Data limits

*What can this data never tell you? Unbalanced history, GSC-only early rows, window overlaps.*


This dataset has several limitations. First, it represents only a limited historical window (mainly 90-day metrics), so it cannot describe long-term changes in page performance. Second, the dataset contains observable search signals rather than confirmed outcomes, so it cannot prove that refreshing a page will improve its performance. Finally, the data may contain missing values and overlapping time windows, which can affect comparisons between pages. Because of these limitations, the model provides decision support rather than causal conclusions.
