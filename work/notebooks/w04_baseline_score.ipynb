## 2. Build the ranked queue (writes the CSV)

*Code the score, rank everything, write work/outputs/baseline_action_score.csv.*

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
import os
import pandas as pd


os.makedirs("work/outputs", exist_ok=True)

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

score = 0

df["baseline_score"] = 0

# Declining trend
df.loc[df["trend_direction"] == "down", "baseline_score"] += 2

# Low CTR
df.loc[df["ctr"] < 0.1, "baseline_score"] += 1

# Poor ranking
df.loc[df["avg_position"] > 20, "baseline_score"] += 1

# Low impressions
df.loc[df["impressions_90d"] < 500, "baseline_score"] += 1

df = df.sort_values("baseline_score", ascending=False)

df.to_csv(
    "work/outputs/baseline_action_score.csv",
    index=False
)

df.head(20)
```
